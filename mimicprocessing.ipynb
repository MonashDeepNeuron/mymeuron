```python
import pandas as pd

edstays_df = pd.read_csv('edstays.csv')

edstays_df.head()

#reads in edstays table and displays first 5 rows
```

```python
import pandas as pd

df_pyx = pd.read_csv('pyxis.csv')

df_pyx.head()

#reads in pyxis table and displays first 5 rows
```

```python
merged_df = pd.merge(df_pyx, edstays_df, on=['subject_id', 'stay_id'], how='inner')

merged_df.head()

#merges pyxis and edstays tabes on subject_id and stay_id
```

```python
merged_df['intime'] = pd.to_datetime(merged_df['intime'])
merged_df['outtime'] = pd.to_datetime(merged_df['outtime'])

#conversts intime and outtime columns to datetime format

merged_df['duration'] = merged_df['outtime'] - merged_df['intime']

merged_df.head()

# operation to get the duration admitted to ED for each subject/stay
```

```python
triage_df = pd.read_csv('triage.csv')

triage_df.head()

#reads in triage table and displays first 5 rows
```

```python
loc_df = pd.merge(merged_df, triage_df, on=['subject_id', 'stay_id'], how='inner')

loc_df.head()

# merges edstays/pyxis dataframe with triage df on subject_id and stay_id
```
