```python
import pke
#import pke library 

extractor = pke.unsupervised.TfIdf()
# current extraction method is the "TfIdf" method from pke library. Others include "KPMiner", "YAKE", "TopicRank".


```

```python
with open('../../complaintscommaseparated.txt', 'r') as file:
    input_data = file.read()
    print(input_data)

extractor.load_document(input=input_data, language='en')
print('no errors I think? :)')
```

```python
#idk how this part works yt :(
# Select keyphrase candidates
extractor.candidate_selection()

# Weight the candidates
extractor.candidate_weighting()

# Get the top N keyphrases
keyphrases = extractor.get_n_best(n=10)

print(keyphrases)
```
